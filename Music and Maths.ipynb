# Square Wave

```python
import math, numpy
from matplotlib import pyplot as plt

x = numpy.linspace(0, 10, 1000)
y = numpy.array([1 if math.floor(2 * t) % 2 == 0 else 0 for t in x])

plt.plot(x,y)
plt.show()
```

```python
# Source : https://stackoverflow.com/questions/12332392/triangle-wave-shaped-array-in-python

phase=-10
length=30 # should be positive
amplitude=10
x=np.arange(0,100,0.1)
def triang(x,phase,length,amplitude):
    alpha=(amplitude)/(length/2)
    return -amplitude/2+amplitude*((x-phase)%length==length/2) \
            +alpha*((x-phase)%(length/2))*((x-phase)%length<=length/2) \
            +(amplitude-alpha*((x-phase)%(length/2)))*((x-phase)%length>length/2)

tr=triang(x,phase,length,amplitude)
plt.plot(tr)
```

```python
# Source : https://www.geeksforgeeks.org/plotting-a-sawtooth-wave-using-matplotlib/


from scipy import signal
import matplotlib.pyplot as plot
import numpy as np
  
t = np.linspace(0, 1, 1000, endpoint=True)
  
# Plot the sawtooth wave
plot.plot(t, signal.sawtooth(2 * np.pi * 5 * t))
  
# Give x, y, title axis label
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.title('Sawtooth Signal')
  
  
# Display
plot.show()
```
